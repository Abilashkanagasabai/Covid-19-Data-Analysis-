### Corona Virus Live Tracking For India 

```python
import pandas as pd

data = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
data
```

```python
data=data.iloc[:-1,1:]
```

```python
data.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states = np.arange(len(data['state_name'])) 
    
```

```python
total_states
```

```python
data['active']
```

```python
max(data['positive'])
```

### Total Positive Cases Based On States In India 

```python

from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,data['positive'], align='center', alpha=0.5,  
                 color=(1,0,0),  
                 edgecolor=(0.5,0.2,0.8))
    
plt.yticks(total_states, data['state_name'])  
plt.xlim(1,max(data['positive'])+100) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()  
```

#### Active Number of Cases based on States

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,data['new_active'], align='center', alpha=0.5,  
                 color=(1,0.5,0),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, data['state_name'])  
plt.xlim(1,max(data['new_active'])+10) 
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()  
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,data['death'], align='center', alpha=0.5,  
                 color=(0,0,1),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, data['state_name'])  
plt.xlim(1,max(data['death'])+10) 
plt.xlabel('Death Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()  
```

```python

#for getting world map
import folium
```

```python
info = pd.read_html("http://www.quickgs.com/latitudinal-and-longitudinal-extents-of-india-indian-states-and-cities/") 

coordinates = pd.DataFrame(info[0])

coordinates.head()
```

```python
data.head()
```

```python
data.head()
covid19=data[["state_name","positive","death","cured","active"]]
```

```python
def data_pre(cord):
    return cord[0:5]
coordinates['Latitude']  = coordinates['Latitude'].apply(data_pre).astype('float')
coordinates['Longitude'] = coordinates['Longitude'].apply(data_pre).astype('float')
coordinates.head()

# cleaning Covid19 Dataframe

# Removing unnecessary rows at the tail and Unnecessary columns


# Renaming Attribute Names for simplicity
covid19.columns = ['State','Total cases','Deaths','Recoveries','Active cases']

covid19.head()
```

```python
final_data = pd.merge(coordinates, covid19, how ='inner', on ='State')
final_data.head()
```

```python
india = folium.Map(location = [20.5937,78.9629],zoom_start=4.5)
#adding to map
for state,lat,long,total_cases,Death,Recov,Active in zip(list(final_data['State']),list(final_data['Latitude']),list(final_data['Longitude']),list(final_data['Total cases']),list(final_data['Deaths']),list(final_data['Recoveries']),list(final_data['Active cases'])):
    #for creating circle marker
    folium.CircleMarker(location = [lat,long],
                       radius = 5,
                       color='red',
                       fill = True,
                       fill_color="red").add_to(india)
    #for creating marker
    folium.Marker(location = [lat,long],
                  # adding information that need to be displayed on popup
                  popup=folium.Popup(('<strong><b>State  : '+state+'</strong> <br>' +
                    '<strong><b>Total Cases : '+str(total_cases)+'</striong><br>' +
                    '<strong><font color= red>Deaths : </font>'+str(Death)+'</striong><br>' +
                    '<strong><font color=green>Recoveries : </font>'+str(Recov)+'</striong><br>' +
                    '<strong><b>Active Cases : '+str(Active)+'</striong>' ),max_width=200)).add_to(india)
    #to show the map
    india
```
